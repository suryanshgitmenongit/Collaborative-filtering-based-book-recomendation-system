```python
#We will display the top 50 rated books, but the initial prerequisite filter is that they must have atleast 250 ratings.
```

```python
ratings
```

```python
#Now, we'll merge the ratings and books dataframes with ISBN as our common column,  because we want a new df that is the union of both these df's
```

```python
books_with_ratings = ratings.merge(books, on = 'ISBN')
```

```python
books_with_ratings
```

```python
#Now, from the new df, we'll find books with minimum 250 ratings and with top 50 ratings
num_rating_df = books_with_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating' : 'num_ratings'})
```

```python
avg_rating_df = books_with_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
#Replacing every book-rating column with the name avg_rating within avg_rating_df
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
```

```python
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['Book-Rating']>=250].sort_values('Book-Rating',ascending=False).head(50)
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Book-Rating','avg_rating']]
```

```python
#Collaborative Filtering Based Recommendation System

x = books_with_ratings.groupby('User-ID').count()['Book-Rating'] > 200
smart_users = x[x].index
```

```python
filtered_rating = books_with_ratings[books_with_ratings['User-ID'].isin(smart_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('How to Kill a Mockingbird')
```
